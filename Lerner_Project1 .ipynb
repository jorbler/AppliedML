## Question 1
If you are rolling five dice, implement a Monte Carlo method to determine if the probability that the sum of the faces showing up is equal to 16 is bigger than the probability that the sum of the faces showing up is equal to 20.

```python
dice = [1, 2, 3, 4, 5, 6]
n = 5
trials = 10000
```

```python
def dice_roll(n, dice, trials):
  spin = [0, 0, 0, 0, 0]
  prob_16 = 0
  prob_20 = 0
  for i in range(trials):
    for i in range(n):
      spin[i] = random.choice(dice)
    if sum(spin) == 16:
      prob_16 += 1
    elif sum(spin) == 20:
      prob_20 += 1
  return (prob_16/trials, prob_20/trials)
```

```python
dice_roll(n, dice, trials)
```

## Question 2
You're about to get on a plane to Boston. You want to know whether it is raining. You call four random friends of yours who live there and ask each one independently if it's raining. The first two of your friends have a 1/5 chance of telling you the truth, and the other two have a 1/3 chance of messing with you by lying. All four friends tell you that "No" it isn't raining. What is the probability that it's raining somewhere in Boston?

In this problem, based on the answers received, the probability of rain in Boston is the same as the probability that at least one of the four friends is lying. We are going to write code to determine this probability experimentally via Monte Carlo simulations.



```python
friend_1 = ['Truth', 'Lie', 'Lie', 'Lie', 'Lie']
friend_2 = ['Truth', 'Lie', 'Lie', 'Lie', 'Lie']
friend_3 = ['Truth', 'Truth', 'Lie']
friend_4 = ['Truth', 'Truth', 'Lie']
```

```python
def rain_prob(trials = 10000):
  lies = 0
  results = [0, 0, 0, 0]
  for i in range(trials):
    results[0] = random.choice(friend_1)
    results[1] = random.choice(friend_2)
    results[2] = random.choice(friend_3)
    results[3] = random.choice(friend_4)
    if 'Lie' in results:
      lies += 1
  return lies/trials

```

```python
rain_prob()
```

##Question 3
Mad Max wants to travel from New York to Dallas by the shortest possible route. He may travel over the routes shown in the table below. Unfortunately, the Wicked Witch can block one road leading out of Atlanta and one road leading out of Nashville. Mad Max will not know which roads have been blocked until he arrives at Atlanta or Nashville. Should Mad Max start toward Atlanta or Nashville?


Route	Length of Route (miles)

New York - Atlanta	866

New York - Nashville	900

Nashville - St. Louis	309

Nashville - New Orleans	532

Atlanta - St. Louis	555

Atlanta - New Orleans	470

St. Louis - Dallas	662

New Orleans - Dallas	505

```python
def atlanta(trials = 1000):
  distance_at = 0
  flip_witch = ['StLouis','NOLA']
  for i in range(trials):
    choice = random.choice(flip_witch)
    if choice == 'StLouis':
      distance_at += (866 + 555 + 662)
    else:
      distance_at += (866 + 470 + 505)
  return distance_at/trials
```

```python
def nashville(trials = 1000):
  distance_nash = 0
  flip_witch = ['StLouis','NOLA']
  for i in range(trials):
    choice = random.choice(flip_witch)
    if choice == 'StLouis':
      distance_nash += (900 + 309 + 662)
    else:
      distance_nash += (900 + 532 + 505)
  return distance_nash/trials
```

```python
def shortest_route(trials):
  if nashville(trials) > atlanta(trials):
    return('Mad Max should head toward Atlanta.')
  else:
    return('Mad Max should head toward Nashville.')
```

```python
shortest_route(1000)
```

## Question 4
Use Monte Carlo simulations to evaluate the probability that a normally distributed variable X, with a mean of 4 and standard deviation of 3, takes values inside the interval between -1 and 2.

```python
size = 10000
loc = 4
scale = 3
x = np.random.normal(loc, scale, size)
```

```python
in_range = 0
for i in range(len(x)):
  if ((x[i] > -1) & (x[i] < 2)):
    in_range += 1
print(in_range/size)
```
